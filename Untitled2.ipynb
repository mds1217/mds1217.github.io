```python
import requests
import pandas as pd
import requests
import xml.etree.ElementTree as ET
from google.colab import userdata
url = "https://www.culture.go.kr"
requests.get(url)
keyID = userdata.get('문체부')
url = f'http://api.kcisa.kr/openapi/service/rest/meta5/getKFCC0502?serviceKey={keyID}&numOfRows=200'
response= requests.get(url).text

root = ET.fromstring(response)
row_dict = {'title':[], 'alternativeTitle':[], 'creator':[], 'regDate':[], 'collectionDb':[],'subjectCategory':[],
            'subjectKeyword':[], 'extent':[], 'description':[], 'spatialCoverage':[], 'temporal':[], 'person':[], 'language':[],
            'sourceTitle':[],'referenceIdentifier':[],'rights':[], 'copyrightOthers':[], 'url':[],'contributor':[]}

for i in root.findall('.body/items/item'):
  for j in i:
    row_dict[j.tag].append(j.text)
df = pd.DataFrame(row_dict)
df.head()
```
